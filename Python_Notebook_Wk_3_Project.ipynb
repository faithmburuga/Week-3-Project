## Renaming Columns

```
# The df3 dataframe has column names that are misspelt. They should be changed to match the other dataframes

df3 = df3.rename(columns={'PRODUTC': 'PRODUCT', 'DATETIME':'DATE_TIME'})
df3
```

```
# The df4 dataframe has column names that are misspelt. They should be changed to match the other dataframes

df4 = df4.rename(columns={'DW_A_NUMBER': 'DW_A_NUMBER_INT', 'DW_B_NUMBER':'DW_B_NUMBER_INT'})
df4
```

```
# The df5 dataframe has column names that are misspelt. They should be changed to match the other dataframes

df5 = df5.rename(columns={'CELLID': 'CELL_ID', 'SIET_ID':'SITE_ID'})
df5
```

# Merging Dataframes

```
# Merge the 3 Telcom datasets since they have similar columns

df_all = pd.concat((df3, df4, df5), axis=0, ignore_index=True)
df_all
```

## Dropping Columns

```
# Dropping unnecessary columns in the table we have merged

df_all = df_all.drop(['COUNTRY_A', 'COUNTRY_B'], axis=1)
df_all
```

```
# Merge the dataframe df1 and df_all on the common column CELL_ID

df_new = pd.merge(df_all, df1, on="CELL_ID")
df_new

# The result shows there are no common values along that column
```

# Analysis

```
# Find out the most used city in the three days
# The mode gives us the most repeated city

df1.VILLES.mode()
```

```
# Get the list of cities in order starting from which was used most to the least used.
# Create a pivot table of counts, by VILLES

df9 = df1.pivot_table(index=['VILLES'], aggfunc='count')
df9

# The output shows the count of the City(Villes)  
```

```
# Arrange the city counts in descending order so we have the order of cities from the highest
# number of counts to the least

df10 = df9.sort_values('AREA',ascending=False)
df10
```
